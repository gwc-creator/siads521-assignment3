# 1. Visualization Technique

The main goal of this dashboard is to tell a clear story about how video game sales vary across years, genres, platforms, publishers, and regions. Instead of using only one chart type, I combined several visualizations so that each one answers a different question.

The dashboard includes the following chart types:

1. **Line chart**  
   This chart shows how total global sales change over time. It is useful for identifying overall trends, peaks, and declines in the market.

2. **Bar charts**  
   I used bar charts to compare total sales by genre and top publishers. Bar charts are a good choice when the goal is to compare categories side by side.

3. **Histogram**  
   The histogram shows the distribution of global sales. This helps reveal whether sales are concentrated among a small number of titles or spread more evenly.

4. **Pie chart**  
   The pie chart displays the regional sales share across North America, Europe, Japan, and Other regions. This provides a quick overview of composition.

5. **Scatter plot**  
   The scatter plot shows the top-selling games by year and global sales, with color used to distinguish genre. This makes it easier to compare individual titles and see how top games differ.

These visualizations complement each other because they show different perspectives of the same filtered data. The line chart shows trend, bar charts show comparison, the histogram shows distribution, the pie chart shows composition, and the scatter plot shows individual observations. Together, they create a more complete dashboard than any single chart alone.

### When these visualizations should be used

This combination works well when a dataset has:
- time information,
- categorical groupings,
- numeric sales measures,
- and multiple dimensions that users may want to compare interactively.

For example, a line chart is especially useful for time-based patterns, while bar charts are better for comparing categories like genre or publisher. A histogram is useful for understanding the shape of a distribution, and a pie chart works for quick proportion comparisons when there are only a few groups. A scatter plot is helpful when comparing individual records.

### Limitations

Each chart also has limitations:
- A line chart is not ideal for category comparisons.
- A pie chart becomes harder to read if there are too many slices.
- A histogram depends on binning choices.
- A scatter plot can become crowded if too many points are plotted at once.

Because of these limitations, combining multiple chart types in one dashboard leads to a better overall user experience.

# 2. Visualization Library

I used **Streamlit** as the dashboard framework and **Plotly** as the charting library.

### Why I chose Streamlit

Streamlit is a Python framework for building data apps and dashboards quickly. It is especially useful for assignments like this because it makes it easy to:
- create a layout,
- add sidebar filters,
- display summary metrics,
- and connect interactions to multiple visualizations.

Another advantage is that Streamlit works well with pandas dataframes and is simple to deploy publicly.

### Why I chose Plotly

Plotly is an interactive visualization library that supports charts such as line charts, bar charts, scatter plots, and pie charts. It is a good choice for dashboards because it allows users to:
- hover over values,
- zoom in,
- and view detailed information interactively.

Plotly also integrates smoothly with Streamlit through `st.plotly_chart()`, which made it easy to build a dashboard with multiple coordinated charts.

### What the dashboard shows

At the top of the dashboard, I included summary metrics:
- total filtered rows,
- number of unique games,
- number of unique publishers,
- and total global sales.

These metrics help users quickly understand the scale of the currently filtered data before looking at the detailed charts.

## Line Chart: Global Sales Over Time

The line chart shows how total global sales change by year under the current filters. This helps identify market growth, peaks, and decline periods.

For example, when the year range is broad, the chart shows the overall rise and fall of the video game market. When the filters are narrowed to specific genres or platforms, the chart becomes more focused and reveals how those groups changed over time.

## Bar Chart: Global Sales by Genre

The genre bar chart compares total global sales across genres. This chart is useful because it makes it easy to see which categories dominate the market under the current filters.

For example, with a broad selection, genres such as Action or Sports may lead in total sales, but after filtering by platform or year range, the ranking can change.

## Bar Chart: Top Publishers by Global Sales

This chart shows the top publishers based on the filtered data. It helps identify which publishers dominate the current market segment selected by the user.

The Top N slider makes this chart more interactive, because users can change how many publishers are shown and compare leaders more clearly.

## Histogram: Distribution of Global Sales

The histogram shows how global sales are distributed across games. This is useful for seeing whether most titles have low sales while only a few titles achieve very high sales.

This chart adds an important distribution view that is different from the category comparisons shown in the bar charts.

## Pie Chart: Regional Sales Share

The regional sales pie chart compares North America, Europe, Japan, and Other regions. This gives a quick view of how the sales composition changes when filters are adjusted.

This chart is especially useful for understanding whether the filtered games are more concentrated in one region or more balanced across regions.

## Scatter Plot: Top-Selling Games

The scatter plot focuses on top-selling games and compares them by year and global sales. I used color to distinguish genre so that the chart adds another layer of comparison.

This chart is helpful because it shifts the focus from aggregated categories to individual titles. It allows the user to see how the biggest-selling games compare with one another.

## Example of coordinated interaction



One of the most important parts of this dashboard is that the filters affect several charts at once.

For example:
- changing the **year range** updates the line chart, bar charts, pie chart, histogram, and scatter plot;
- selecting only a few **genres** changes category comparisons and also changes the sales distribution and regional composition;
- selecting specific **platforms** narrows the dashboard to a more focused market segment;
- changing **Top N** changes how many publishers or top titles are shown.

This coordinated filtering makes the dashboard more useful than a collection of unrelated charts.
